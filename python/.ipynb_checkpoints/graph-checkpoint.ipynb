```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from PIL import Image
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv('ANALYSIS/image.csv',index_col=0)
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data['size_of_file'], data['encrpyt_time'],'ro',label='Encryption') 
plt.plot(data['size_of_file'], data['decrypt_time'],'bo',label='Decryption')  
plt.legend()
plt.xlabel('Size of Image (bytes)') 
plt.ylabel('Time (seconds)') 
plt.title('Encryption-Decryption') 
# plt.show() 
plt.savefig('ANALYSIS/image_performance.png')
```

```python
data = pd.read_csv('ANALYSIS/video.csv')
figure(num=None, figsize=(4.5, 3.2), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data['size_of_file'], data['encrpyt_time'],'ro',label='Encryption') 
plt.plot(data['size_of_file'], data['decrypt_time'],'bo',label='Decryption') 
  
plt.xlabel('Size of Video (bytes)') 
plt.ylabel('Time (seconds)') 
  
# giving a title to my graph 
plt.title('Encryption-Decryption') 
  
# function to show the plot 
# plt.show() 
plt.savefig('ANALYSIS/video_performance.png')
```

```python
data = pd.read_csv('ANALYSIS/audio.csv')
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
data.sort_values(by="encrpyt_time")
plt.plot(data['size_of_file'], data['encrpyt_time'],'ro',label='Encryption') 
plt.plot(data['size_of_file'], data['decrypt_time'],'bo',label='Decryption') 
plt.legend()

  
plt.xlabel('Size of Audio (bytes)') 
plt.ylabel('Time (seconds)') 
plt.title('Encryption-Decryption') 
# plt.show() 
plt.savefig('ANALYSIS/audio_performance.png')
```
